```python
def vigenere(s : str, a : int, b : int):
    v = True
    for i in range(len(s)):
        tmp = ord(s[i]) - ord('a')   # mettre les char entre 0 et 25
        if v:
            tmp = (tmp + a) % 26
            v = False
        else:
            tmp = (tmp + b) % 26
            v = True
        print(chr(tmp + 63),end=" ")
        #print(tmp)
    
```

```python
vigenere("deschercheurstentent",2,19)
```

```python
def chiffrement_affine(s: int, a : int, b : int):
    for i in range(len(s)):
        tmp = ord(s[i]) - ord('a')
        #print(tmp,end=" ")
        tmp = (a*tmp + b) % 26
        print(chr(tmp+65),end=" ")
```
