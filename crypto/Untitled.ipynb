```python
def vigenere(s:str, k : str):
    j = 0
    for i in range(len(s)):
        if j == len(k): # i % len(k)
            j = 0
        tmp = ord(s[i]) - ord('a')
        tmp2 = ord(k[j]) - ord('a')
        res = (tmp + tmp2) % 26
        #print(res,end=" ")
        print(chr(res+ord('a')),end=" ")
        j += 1
```

```python
def chiffrement_affine(s: int, a : int, b : int):
    for i in range(len(s)):
        tmp = ord(s[i]) - ord('a')
        #print(tmp,end=" ")
        tmp = ((a*tmp) + b) % 26
        print(chr(tmp+65),end=" ")
```

```python
chiffrement_affine("chiffrementaffine",5,8)
```

```python
def dechiffrement_affine(s : str, a : int, b : int):
    inv_a = pow(a,-1,26)
    print(inv_a)
    for i in range(len(s)):
        tmp  = ord(s[i]) - ord("a")
        x = tmp - b % 26
        tmp = (inv_a * x) % 26
        print(chr(tmp+65),end=" ")
```

```python
def freq(s : str):
    dico = {}
    cpt = 0
    for i in range(len(s)):
        cpt = 0
        tmp = s[i]
        if tmp not in dico:
            for j in range(len(s)):
                if tmp == s[j]:
                    cpt += 1
            dico[tmp] = cpt
    return dict(sorted(dico.items(), key=lambda item: item[1], reverse = True))
```

```python
def occ(s):
    dico = {}
    for i in range(len(s)):
        cpt = 0
        for j in range(len(s)):
            if s[i] == s[j]:
                cpt += 1
        if s[i] not in dico:
            dico.update({s[i]:cpt})
    return dico
```
