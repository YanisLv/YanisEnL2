```python
#1
def vigenere(s:str, c : str):
    for i in range(len(s)):
        tmp = ord(s[i])-ord('A')
        tmp2= ord(c[i%len(c)])-ord('A')
        res = (tmp+tmp2) % 26
        print(chr(res+ord('A')),end=" ")
vigenere("CETEXAMENESTFACILE","KEY")

def dech_vigenere(s:str,c:str):
    for i in range(len(s)):
        tmp1 = ord(s[i])-ord('A')
        tmp2 = ord(c[i%len(c)])-ord('A')
        res = (tmp1-tmp2) % 26
        print(chr(res+ord('A')),end=" ")

```

```python
#3
def freq(s:str):
    dico={}
    for i in range(len(s)):
        cpt = 0
        for j in range(len(s)):
            if s[i] == s[j]:
                cpt += 1
        if s[i] not in dico:
            dico.update({s[i]:cpt})
    return dico
print(freq("AHOXUREZZOSHIPSDHHDSHFLCPQAFPAMMDUF"))      #CE TEXTE EST CHIFFRÉ
print('\n')
print(freq("BTEDHGHAOSRLNRQMLVXCAKWVGSGATSRICOI"))
print('\n')
print(freq("CWOQKPUEGEJKHHTGOGPVGUVWPRGWRNWUUWT"))
dech_vigenere("EOJRVWVWXSKDGHULOWJHXDRUMFRQVVVVVCDPXIEETASRNZRJKOEGXOWITWIHWIEKHADHXGKGXAFQMSIGTBJOTQCDLGVVNDVUBSLUXOCDLWVQGSVWMCLWESWIHFKGXQVWMSTOTGJHXGKGXZVPISTKXFUHFCEWXF","TORD")
#CLÉ = TORD
```
