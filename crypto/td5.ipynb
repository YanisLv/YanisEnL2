```python
def nb(a : int):
    i = 0
    while a > 0:
        if a % 2 == 1:
            i += 1
        a = a//2
    return i-1 
```

```python
from math import gcd
def roh_roh_pollard(n):
    if n % 2 == 0: return 2
    if n == 1: return 1
    def f(x):
        return x*x+1
    x, y, d = 2, 2, 1
    while d==1:
        x = f(x) % n
        y = f(f(y)) % n
        #print (x,y)
        d = gcd(x-y, n) #pgcd(x-y, n)
        if d == n:
            return roh_roh_pollard(n)
        
    return d
```

```python
from math import gcd
def rp(n):
    if n % 2 == 0: return 2
    if n == 1 :return 1
    
    def f(x):
        return pow(x,2,n)
    x,y,d = 2,2,1
    while d==1: #cas 1
        x = f(x)
        y = f(f(y))
        d = gcd(x-y,n)
        if d == n:  #cas 2
            return roh_roh_pollard(n)
    #cas 3
    return d
```
